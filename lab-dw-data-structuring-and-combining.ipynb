In this challenge, we will be working with the customer data from an insurance company, as we did in the two previous labs. The data can be found here:

But this time, we got new data, which can be found in the following 2 CSV files located at the links below.

- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv

Note that you'll need to clean and format the new data.

Observation:
- One option is to first combine the three datasets and then apply the cleaning function to the new combined dataset
- Another option would be to read the clean file you saved in the previous lab, and just clean the two new files and concatenate the three clean datasets

```python
import pandas as pd
url1 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv'
url2 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv'
url3 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv'
customer1_df = pd.read_csv(url1)
customer2_df = pd.read_csv(url2)
customer3_df = pd.read_csv(url3)
```

```python
customer1_df.rename(columns= {'ST': 'state'}, inplace=True)
customer1_df.columns = customer1_df.columns.str.lower()
customer1_df.columns = customer1_df.columns.str.replace(' ', '_')
customer1_df
```

```python
customer2_df.rename(columns= {'ST': 'state'}, inplace=True)
customer2_df.columns = customer2_df.columns.str.lower()
customer2_df.columns = customer2_df.columns.str.replace(' ', '_')
customer2_df
```

```python
customer3_df.columns = customer3_df.columns.str.lower()
customer3_df.columns = customer3_df.columns.str.replace(' ', '_')
customer3_df
```

```python
update_df = pd.concat([customer1_df, customer2_df, customer3_df], axis=0)
update_df
```

```python
update_df.columns
```

```python
update_df['state'].unique()
```

```python
update_df['state'] = update_df['state'].replace({
    'AZ': 'Arizona',
    'WA': 'Washington',  
    'Cali': 'California'      
})
```

```python
update_df['gender'].unique()
```

```python
update_df['gender'] = update_df['gender'].replace({
    'Femal': 'F',
    'female': 'F',  
    'Male': 'M'      
})
```

```python
update_df['education'].unique()
```

```python
update_df['education'] = update_df['education'].replace({
    'Bachelors': 'Bachelor'   
})
```

```python
update_df['vehicle_class'].unique()
```

```python
update_df['vehicle_class'] = update_df['vehicle_class'].replace({
    'Luxury SUV': 'Luxury',
    'Sports Car': 'Luxury',  
    'Luxury Car': 'Luxury'      
})
```

```python
update_df['customer_lifetime_value'] = update_df['customer_lifetime_value'].str.replace('%', '').astype(float)
```

```python
def extract_middle_digit(value):
    if pd.isna(value):
        return
    if isinstance(value, int):
        return value
    return int(value.split("/")[1])

update_df['number_of_open_complaints'] = update_df['number_of_open_complaints'].apply(extract_middle_digit)
```

```python
update_df
```

```python
update_df['customer_lifetime_value'] = pd.to_numeric(update_df['customer_lifetime_value'])
```

```python
update_df.isna().any()
```

```python
update_df.isna().sum()
```

```python
update_df = update_df.dropna(how='all')
update_df
```

```python
update_df.isna().sum()
```

```python
update_df["gender"].fillna('Prefer not to say')
```

```python
update_df['customer_lifetime_value'].fillna(update_df['customer_lifetime_value'].mean())
```

```python
update_df
```

```python
update_df.duplicated().sum()
```

```python
update_df_clean = update_df
```

```python
update_df_clean = update_df_clean.drop_duplicates()
update_df_clean
```

```python
update_df_clean.reset_index().tail()
```

```python
update_df_clean.to_csv("Customer123_Data_Clean.csv")
```
